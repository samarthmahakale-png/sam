```python
df['Department'].fillna(df['Department'].mode()[0],inplace=True)
df['Position'].fillna(df['Position'].mode()[0],inplace=True)
df['State'].fillna(df['State'].mode()[0],inplace=True)
df['Experience_Years'].fillna(df['Experience_Years'].mode()[0],inplace=True)
df['Salary'].fillna(df['Salary'].mode()[0],inplace=True)
df['Performance_Rating'].fillna(df['Performance_Rating'].mode()[0],inplace=True)
df['Joining_Date'].fillna(df['Joining_Date'].mode()[0],inplace=True)
df['Projects_Completed'].fillna(df['Projects_Completed'].mode()[0],inplace=True)
df['Overtime_Hours'].fillna(df['Overtime_Hours'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Salary']=df['Salary'].replace([np.inf,-np.inf,np.nan])
```

```python
df['Joining_Date']=pd.to_datetime(df['Joining_Date'].fillna(df['Joining_Date']))
df.info()
```

```python
df['Age']=df['Age'].astype(int)
df['Salary']=df['Salary'].astype(int)
df['Performance_Rating']=df['Performance_Rating'].round(1)
columns_to_drop = ['Experience_Years','State','Education','Joining_Date','Projects_Completed']
```

```python
import matplotlib.pyplot as plt
dept = df['Department'].value_counts()
plt.bar(dept.index,dept.values)
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Department WISE employees')
plt.show()
```

```python
gender = df['Gender'].value_counts()
plt.pie(gender.values, labels=gender.index, autopct='%1.1f%%')
plt.title('Gender WISE employees')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='cool', fmt='.3f', linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
Salary_exp = df.groupby("Experience_Years")["Salary"].mean()
plt.plot(Salary_exp.index,
         Salary_exp.values,
         marker="o")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(Salary_exp.index, Salary_exp.values, marker='o', linestyle='-')
plt.title('Average Salary by Experience (Line Plot)')
plt.xlabel('Experience Years')
plt.ylabel('Average Salary')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
average_salary_by_department = df.groupby('Department')['Salary'].mean()

plt.figure(figsize=(12, 7))
plt.bar(average_salary_by_department.index, average_salary_by_department.values, color='skyblue')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Average Salary by Department (Bar Chart)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=10, kde=True)
plt.title('Distribution of Employee Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
department_counts = df['Department'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Percentage of Employees in Each Department')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Experience_Years', y='Salary', data=df)
plt.title('Relationship between Salary and Experience Years')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
