```python
df['Department'].fillna(df['Department'].mode()[0],inplace=True)
df['Position'].fillna(df['Position'].mode()[0],inplace=True)
```

```python
df['State'].fillna(df['State'].mode()[0],inplace=True)
df['Experience_Years'].fillna(df['Experience_Years'].mode()[0],inplace=True)
df['Salary'].fillna(df['Salary'].mode()[0],inplace=True)
df['Performance_Rating'].fillna(df['Performance_Rating'].mode()[0],inplace=True)
df['Joining_Date'].fillna(df['Joining_Date'].mode()[0],inplace=True)
df['Projects_Completed'].fillna(df['Projects_Completed'].mode()[0],inplace=True)
df['Overtime_Hours'].fillna(df['Overtime_Hours'].mode()[0],inplace=True)
```

```python
df.rename(columns={'Salary':'salary RS'},inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler()

# Replace infinite values with NaN
df['salary RS'] = df['salary RS'].replace([np.inf, -np.inf], np.nan)

# Fill any newly created NaN values (from inf replacement) with the mode
df['salary RS'] = df['salary RS'].fillna(df['salary RS'].mode()[0])

df["salary RS"] = scaler.fit_transform(df[["salary RS"]])
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Experience_Years'], df['Projects_Completed'], color='blue', alpha=0.7, edgecolors='k')

# Add labels and title
plt.title('Experience Years vs. Projects Completed', fontsize=14)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Projects Completed', fontsize=12)

# Add a grid for readability
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```
