```python
# @title
# Import Libraries
import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_csv("Cleaned_Indian_Employee_Data.csv")

# Display first 5 rows
print(df.head())

# Check dataset information
print(df.info())

# Check missing values
print(df.isnull().sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

# Handle infinite values in 'Experience_Years' by replacing them with NaN
df["Experience_Years"] = df["Experience_Years"].replace([np.inf, -np.inf], np.nan)
# Now fill NaN values (including those converted from inf) with the median
df["Experience_Years"] = df["Experience_Years"].fillna(df["Experience_Years"].median())

# Fill missing categorical values
df["Department"] = df["Department"].fillna(df["Department"].mode()[0])
df["City"] = df["City"].fillna(df["City"].mode()[0])

# Remove rows with remaining missing values
df.dropna(inplace=True)

# Convert data types
df["Salary"] = df["Salary"].astype(int)
df["Experience_Years"] = df["Experience_Years"].astype(int)

# Save cleaned dataset
df.to_csv("Cleaned_Indian_Employee_Data.csv", index=False)

print("Dataset cleaned successfully!")
```

```python
print("Minimum Performance_Rating in df:", df["Performance_Rating"].min())
print("Maximum Performance_Rating in df:", df["Performance_Rating"].max())
```

```python
# Create Target Variable 'promotion' based on 'Performance_Rating'
df["promotion"] = (df["Performance_Rating"] >= 3).astype(int)

eligible = df[df["promotion"] == 1]

print("=== Eligible for Promotion ===")
print(eligible[["Employee_ID", "Name", "Performance_Rating"]])
```

```python
# Function to calculate increment based on performance rating
def calculate_increment(performance_rating):
    if 3.5 <= performance_rating < 4.2:
        return 2.5   # 2.5% increment

    elif 4.2 <= performance_rating < 5.0:
        return 7.5   # 7.5% increment

    elif 5.0 <= performance_rating <= 5.5:
        return 15.0  # 15% increment

    else:
        return 0.0   # No increment


# Apply increment calculation
df["Increment_Percentage"] = df["Performance_Rating"].apply(calculate_increment)

# Calculate increment amount
df["Increment_Amount"] = df["Salary"] * df["Increment_Percentage"] / 100

# Calculate new salary
df["New_Salary"] = df["Salary"] + df["Increment_Amount"]


# Add eligibility column
df["Increment_Eligible_Binary"] = (df["Increment_Percentage"] > 0).astype(int)


# Filter eligible employees
eligible_for_increment = df[df["Increment_Eligible_Binary"] == 1]

# Display eligible employees
print(eligible_for_increment)
```
